```python
data_pipez = ['da', 'db', 'dc']
model_pipez = ['mx', 'my', 'mz']
# data_pipez = np.array(['a', 'b', 'c'])
# model_pipez = np.array(['x', 'y', 'z'])

data_pipez = [np.random.randint(0,100, size=2), np.random.randint(0,100, size=2), np.random.randint(0,100, size=2)]
model_pipez = [lambda x: print('identity: ', x), 
               lambda x: print( 'square: ', x**2), 
               lambda x: print('log: ', np.log(1/x))]

print( data_pipez, "\t", model_pipez)


# d_, m_ = np.meshgrid(data_pipez, model_pipez)
d_, m_ = np.meshgrid(range(len(data_pipez)), range(len(model_pipez)))
#print(d_, "\n", m_)
```

```python
O_ = []
for x,y in zip(d_, m_):
#     print(x, "\t", y)
    for i, j in zip(x,y):
        O_.append( (data_pipez[i], model_pipez[j]) )
#         print( f"{i},{j}")
#         O_.append( (i,j))
        
O_
```

```python
for d, m in O_:
    print(d)
    np.vectorize(m)(d)
    print("-----")
```

```python
for d, m in O_:
    print(d)
    for i in d:
        m(i)
    print("-----")
```
