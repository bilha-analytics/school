# 2. Load fundus Images

```python
fundus_imagez = []
for fp in glob.glob( f"{DATA_DIR}/*.ppm"):
    img = utilz.Image.fetch_and_resize_image(fp, None)     
    fname = os.path.splitext( os.path.basename(fp) )[0]
    stats = utilz.FileIO.image_file_parser(fp)
    fundus_imagez.append( (img, fname, stats) )
    
    
utilz.Image.plot_images_list( [f[0] for f in fundus_imagez], titlez=[t[1]for t in fundus_imagez] )
```

# 3. Different Handcrafterd Features

```python
import extract 

fundus_color_channels_mapper = extract.FundusColorChannelz()

fcolz = fundus_color_channels_mapper.transform( [f[0] for f in fundus_imagez])

print( fcolz[0].shape)

for fi in range(len(fcolz)): 
    tz = [ f"{fundus_imagez[fi][1][-3:]}_{x}" for x in ['vessel', 'green', 'red', 'blue'] ]
    r = fcolz[fi]
    _,_,c = r.shape 
    utilz.Image.plot_images_list( [-r[:,:,i] for i in range(c)]+[ -r[:,:,1:]], titlez=tz+['all'] , nc=c+1, cmap='gray')
```

```python

def _disp_bluez(bluez, show_hist=True):
    n = len(bluez)
    utilz.Image.plot_images_list( [b.astype('uint8') for b in bluez ], cmap='gray')
    if show_hist:
        for i in range( n ):
            plt.subplot(1, n, (i+1))
            plt.hist(bluez[i].flatten() )
        plt.show();
        plt.clf()
    for b in bluez:
        print( f"min: {b.min()}, max={b.max()}")

        
## origi
origi_bluez = [ f[0][:,:,-1] for f in fundus_imagez]
_disp_bluez(origi_bluez)
## remapped
remapped_bluez = [ f[:,:,-1] for f in fcolz]
_disp_bluez(remapped_bluez)
```

```python
def _update_blue(blu, dig=False):
    if dig:
        return blu.copy()
    
    blu = blu.copy() 
    blu[ blu != 1] = 0
    return blu


## origi
origi_bluez_update = [ _update_blue( f[0][:,:,-1] ) for f in fundus_imagez]
_disp_bluez(origi_bluez_update, show_hist=False)
## remapped
remapped_bluez_update = [ _update_blue( f[:,:,-1] , dig=True) for f in fcolz]
_disp_bluez(remapped_bluez_update, show_hist=False)
```

```python
def _nfimg(img):
    x, y, c = img.shape
    print("\nImage:")
    for i in range(c):
        print( f"\tChannel: {i} min: {img[:,:,i].min()}, max: {img[:,:,i].max()}")
    
## fullrammped image
vesall_remapped = [ f[:,:,:-1]  for f in fcolz]
_disp_bluez(vesall_remapped, show_hist=False)

all_remapped = [ f[:,:,1:]  for f in fcolz]
_disp_bluez(all_remapped, show_hist=False)


## the numbers 
_ = [ _nfimg(f) for f in fcolz]
```

```python
 def mask_out_border(img):
    border_mask =  img.copy()[:,:,2] 
    bmax = border_mask.max() 
    bmin = border_mask.min()
#     border_mask[ border_mask != bmin ] = 255
#     border_mask[ border_mask == bmin ] = -1
#     border_mask[ border_mask == bmax ] = 1
    border_mask[ border_mask != 0 ] = 200
    print(f"\nBorder_Mask:\tmin: {border_mask.min()}, max: {border_mask.max()}")    
    return border_mask

def update_masked(img):
    border = mask_out_border(img)
    outie = img.copy()
    _,_,c = outie.shape 
    outie = [ outie[:,:,i]*border for i in range(c)]
    return np.dstack(outie)

masked = [ mask_out_border(f)  for f in fcolz]
_disp_bluez(masked, show_hist=False)


dmasked = [ update_masked(f)  for f in fcolz]
_disp_bluez(dmasked, show_hist=False)


n = len( masked )
for i in range( n ):
    plt.subplot(1, n, (i+1))    
    plt.hist( masked[i].flatten() ); 
```

```python
## cropping
def _crop(im, f=1):
    o = im.copy()
    x, y = o.shape[0], o.shape[1] 
    ox, oy = int(x*f), int(x*f)
    ix, iy = (x-ox)//2, (y-oy)//2
    o = o[ix:ix+ox, iy:iy+oy, :]
    return o


cropped = [ _crop(f[0])  for f in fundus_imagez]
_disp_bluez(cropped, show_hist=False)
```

# Local Binary Pattern et. al. 

```python
from skimage.feature import local_binary_pattern

def v_mask(v):
    o = v.copy()
    omax = o.min()
#     print( omax )
#     o[ (o >= 200) & (o <= 255) ] = 0
    o[ o >= omax*0.85 ] = 1
    o[ o < omax*.85 ] = 0
    return o*255
def _deveined_f(img):
    v = v_mask(img[:,:,0]) 
    g = img[:,:,1]
    o = g*v
    return o

lbp_radius = 1
lbp_k = 8*lbp_radius
lbp_method = 'uniform'
lbp_imgz = [ local_binary_pattern(_deveined_f(f), lbp_k, lbp_radius, lbp_method) for f in fcolz]


# vessels
_disp_bluez([f[:,:,0] for f in fcolz])
_disp_bluez([-v_mask(f[:,:,0]) for f in fcolz])

# deveined LBP
_disp_bluez(lbp_imgz)
```
